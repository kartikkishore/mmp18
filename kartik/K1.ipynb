```python
COUNT = train.groupby(['AppVersion']).count()['HasDetections'].to_frame()
```

```python
SUM = train.groupby(['AppVersion']).sum()['HasDetections'].to_frame()
```

```python
faultyAppVer = pd.merge(left=COUNT, right=SUM,on='AppVersion')
faultyAppVer['hasDetAppVer_Percentage'] = (faultyAppVer['HasDetections_y']/faultyAppVer['HasDetections_x'])*100
faultyAppVer.drop(columns=['HasDetections_x','HasDetections_y'], inplace=True)
sns.barplot(x='hasDetAppVer_Percentage', data=faultyAppVer, )
```
